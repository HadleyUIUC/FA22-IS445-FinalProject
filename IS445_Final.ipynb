# Scatter Plot: Magnitude vs Depth of Earthquake

```python
#Magnitude vs Depth of Earthquake
fig, ax = plt.subplots()
x = USGS_30_day['depth']
y = USGS_30_day['mag']
ax.scatter(x, y)

ax.set_xlabel('Depth')
ax.set_ylabel('Magnitude')

plt.show()
```
